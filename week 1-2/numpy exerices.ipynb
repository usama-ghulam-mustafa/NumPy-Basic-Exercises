```python
import matplotlib.pyplot as  plt
import numpy as np
```

```python
#exercise 1: loading an image and printing its shape,dtype,max and min
img = plt.imread(r"D:\ML and AI\Data sets\Cats and Dogs\train\dogs\dog_8.jpg")
print(f"{img.shape}")
print(f"{img.dtype}")
print(f"{img.max()}")
print(f"{img.min()}")
```

```python
#exercise 2: separating an image into 3 different RGB channels
r_channel = img[:,:,0]
g_channel = img[:,:,1]
b_channel = img[:,:,2]

print(r_channel)
print(g_channel)
print(b_channel)
```

```python
#Exercise 2: displaying the channels
fig, ax = plt.subplots(1,3, figsize=(9,3))
ax[0].imshow(r_channel, cmap='gray')
ax[1].imshow(g_channel, cmap='gray')
ax[2].imshow(b_channel, cmap='gray')
```

```python
#Exercise 3:
gray_img = plt.imshow(r_channel*0.2989 + 0.1140*b_channel + 0.5870*g_channel , cmap="gray")
print(gray_img)
```

```python
#Exercise 4: cropping the image 100 by 100 pixel from the center
plt.imshow(img[490:590, 910:1010,:])
```

```python
#Exercise 5: horizontally flipping the image
plt.imshow(img[::,::-1])
```

```python
#Exercise 6: Brightening the image by adding 50 to every pixel
plt.imshow(np.clip(img.astype(np.int16)+50, a_min=0, a_max=255).astype(np.uint8))
```

```python
#Exercise 7: pixel value histogram for RGB channel on same graph
plt.figure(figsize=(10,10))
plt.hist(r_channel.flatten(), color="Red", alpha=0.3, bins=256)
plt.hist(g_channel.flatten(), color="Green", alpha=0.3, bins=25)
plt.hist(b_channel.flatten(), color="Blue", alpha=0.3, bins=256)
plt.show
```

```python
#Exercise 8:
gray_img_arr = gray_img.get_array()
empty_arr = np.zeros_like(gray_img_arr)
mask = gray_img_arr > 128
empty_arr[mask] = 255
plt.imshow(empty_arr, cmap="gray")
```

```python
#Exercise 9: normalizing the image
normalized_img = img*(1./255)
plt.imshow(normalized_img)
```

```python
#Exercise 10: subplot of original image and grayscale image
fig, ax = plt.subplots(1,2,figsize=(8,4))
ax[0].imshow(gray_img_arr, cmap="gray")
ax[1].imshow(img)
```
